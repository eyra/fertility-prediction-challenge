```python
import os.path
import pandas as pd
from joblib import dump, load
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_csv(os.path.join("..", "data", "fake_data.csv"), encoding="latin-1", encoding_errors="replace")
```

```python
outcomes = pd.read_csv(os.path.join("..", "data", "fake_data_ground_truth.csv"), encoding="latin-1", encoding_errors="replace")
```

```python
regr = linear_model.LinearRegression()
regr.fit(data[["fake_value"]], outcomes[["outcome"]])
```

```python
dump(regr, os.path.join("..", "models", "model.joblib"))
```
